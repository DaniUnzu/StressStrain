```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import minimize
```

```python
df = pd.read_csv("datafile.csv", header = None)
df.columns = ["S0", "k", "n", "m", "E*", "m6", "E", "S"]
```

```python
df.head()
```

```python
df.shape
```

```python
df_first = df.iloc[:240, :]
```

```python
plt.plot(df_first["E"], df_first["S"])
```

```python

# Define the viscoplastic law
def viscoplastic_law(params, epsilon, epsilon_dot, sigma_exp, sigma0):
    k, n, m, m6 = params
    sigma_pred = k * epsilon**n * epsilon_dot**m + sigma0 * np.exp(m6*epsilon)
    return np.mean((sigma_pred - sigma_exp)**2)


def optimise_law(df_curve):
    
    error_at_each_step = []
    def callback_func(params):
        error = viscoplastic_law(params, epsilon, epsilon_dot, sigma_exp, sigma0)
        error_at_each_step.append(error)
    
    # Initialize parameters
    params_init = np.array([100, 1, 1, 0])  # Initial guess for k, n, m

    epsilon = df_curve["E"].values
    epsilon_dot = df_curve["E*"].values
    sigma_exp = df_curve["S"].values
    sigma0 = df_curve["S0"].values

    # Perform gradient descent using SciPy's minimize function
    result = minimize(viscoplastic_law, params_init, args=(epsilon, epsilon_dot, sigma_exp, sigma0), method='L-BFGS-B', callback=callback_func)

    # Extract optimal parameters
    optimal_params = result.x
    k_opt, n_opt, m_opt, m6_opt = optimal_params
    
    plt.plot(error_at_each_step)
    
    print("FINAL ERROR", error_at_each_step[-1])
    # Print the optimal parameters
    print("Optimal parameters:")
    print("k =", k_opt)
    print("n =", n_opt)
    print("m =", m_opt)
    print("m6 =", m6_opt)
    
    return optimal_params
```

```python
k_opt, n_opt, m_opt, m6_opt = optimise_law(df_first)
```

```python
def validate(df_curve, k_opt, n_opt, m_opt, m6_opt):
    row1 = df_curve.iloc[0]
    sigma0 = row1["S0"]
    
    predicted_stress = sigma0 + k_opt * df_curve["E"]**n_opt * df_curve["E*"]**m_opt * np.exp(m6_opt*epsilon)

    sf =  sigma0 + row1["k"] * df_curve["E"]**row1["n"] * df_curve["E*"]**row1["m"]
    
    plt.scatter(df_curve["E"], predicted_stress, label = "predicted", s = 1, c = "green")
    plt.plot(df_curve["E"], df_curve["S"], label = "S")
    plt.plot(df_curve["E"], sf, label = "Sf")
    plt.legend()
    plt.show()
```

```python
validate(df_first, k_opt, n_opt, m_opt, m6_opt)
```

```python
len(df)/240
```

```python
def compute_curves(df):
    results = []
    for i in range(len(df)//240):
        df_curve = df[240 * i: 240 * (i+1)]
        k_opt, n_opt, m_opt, m6_opt = optimise_law(df_curve)
        error = viscoplastic_law((k_opt, n_opt, m_opt, m6_opt), df_curve["E"].values, df_curve["E*"].values, df_curve["S"].values, df_curve["S0"].values)
        results.append({'k_opt': k_opt, 'n_opt': n_opt, 'm_opt': m_opt, 'm6_opt': m6_opt, 'error': error})
    return pd.DataFrame(results)
```

```python
results = compute_curves(df)
```

```python
results["error"].max()
```

```python
plt.figure(figsize=(8, 6))
plt.boxplot(results['error'], vert=False, patch_artist=True, showfliers=True)
plt.title('Box Plot of Error with Outliers')
plt.xlabel('Error')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(results['error'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Error')
plt.xlabel('Error')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
def validate_from_index(df, curve_index):
    df_curve = df[240*curve_index: 240*(curve_index+1)]
    k_opt, n_opt, m_opt, m6_opt = optimise_law(df_curve)
    validate(df_curve, k_opt, n_opt, m_opt, m6_opt)
```

```python
validate_from_index(df, 35)
```
