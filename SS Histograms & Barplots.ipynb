```python
import pandas as pd
import matplotlib.pyplot as plt
import math
import seaborn as sns

# Load the dataset
def load_dataset(file_path):
    df = pd.read_csv(file_path)
    return df

# Preview the dataset
def preview_dataset(df, n=5):
    return df.head(n)

# Check for missing values
def check_missing_values(df):
    return df.isnull().sum()

# Perform summary statistics
def summary_statistics(df):
    return df.describe()

# Visualize the distribution of each numerical column using histograms
def plot_histograms(df, bins=10, cols_per_row=4):
    num_cols = df.select_dtypes(include=['number']).columns
    num_plots = len(num_cols)
    num_rows = math.ceil(num_plots / cols_per_row)

    fig, axs = plt.subplots(nrows=num_rows, ncols=cols_per_row, figsize=(15, 5 * num_rows))

    for i, col in enumerate(num_cols):
        row = i // cols_per_row
        col_idx = i % cols_per_row

        ax = axs[row, col_idx]
        df[col].plot(kind='hist', bins=bins, ax=ax)
        ax.set_title(f'Histogram of {col}')
        ax.set_xlabel(col)
        ax.set_ylabel('Frequency')

    # Remove empty subplots
    for i in range(num_plots, num_rows * cols_per_row):
        fig.delaxes(axs.flatten()[i])
    
    plt.tight_layout()
    plt.show()


# Visualize the relationship between numerical columns and 'Behavior' column using box plots
def plot_boxplots_grouped_by_behavior(df, behavior_col='Behavior'):
    num_cols = df.select_dtypes(include=['number']).columns
    for col in num_cols:
        plt.figure()
        df.boxplot(column=col, by=behavior_col)
        plt.title(f'Boxplot of {col} grouped by {behavior_col}')
        plt.xlabel(behavior_col)
        plt.ylabel(col)
        
#         if order:
#             plt.xticks(range(1, len(order) + 1), order)
        
        plt.show()      
        
    
# Example usage
file_path = r"C:\Users\dunzu\OneDrive\Desktop\Project\Dataset.csv"
df = load_dataset(file_path)
print(preview_dataset(df))
print(check_missing_values(df))
print(summary_statistics(df))
plot_histograms(df)
plot_boxplots_grouped_by_behavior(df)
```
